```python
from google.colab.patches import cv2_imshow
import cv2
image = cv2.imread("images.png")
cv2_imshow(image)
```

```python
grey = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2_imshow(grey)
```

```python
invert = cv2.bitwise_not(grey)
cv2_imshow(invert)
```

```python
blur = cv2.GaussianBlur(invert, (21,21),0)
cv2_imshow(blur)
```

```python
invertblur = cv2.bitwise_not(blur)
cv2_imshow(invertblur)
```

```python
sketch = cv2.divide(grey,invertblur,scale=255.0)
cv2_imshow(sketch)
cv2.imwrite("images.png",sketch)
```
